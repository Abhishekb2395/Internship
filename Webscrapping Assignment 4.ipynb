Q1 : Scrape the details of most viewed videos on YouTube from Wikipedia:
 Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
 You need to find following details:
 A) Rank
 B) Name
 C) Artist
 D) Upload date
 E) Views

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
import requests
from bs4 import BeautifulSoup
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome(r"chromedriver.exe")
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
time.sleep(2)
driver.maximize_window()
```

```python
Rank = []
Name = []
Artist = []
Upload_date = []
Views = []
```

```python
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]"):
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append("-")
    
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]"):
        Name.append(i.text)
except NoSuchElementException:
    Name.append("-")
    
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]"):
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("-")
    
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]"):
        Upload_date.append(i.text)
except NoSuchElementException:
    Upload_date.append("-")
    
try:
    for i in driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]"):
        Views.append(i.text)
except NoSuchElementException:
    Views.append("-")
```

```python
len(Rank),len(Name),len(Artist),len(Upload_date),len(Views)
```

```python
df = pd.DataFrame({"Rank":Rank,"Name":Name,"Artist":Artist,"Upload Date":Upload_date,"Views(Billions)":Views})

df.Name = df.Name.apply(lambda x:x[:-4].strip('"'))
df
```

```python
Dataset_name = []
Data_type = []
Task = []
Attribute_type = []
No_of_instances = []
No_of_attribute = []
Year = []
```

```python
for i in urls:
    driver.get(i)
    time.sleep(3)
    
    try:
        dataset_name = driver.find_element(By.XPATH,'//*[@class="flex flex-col"]/div[1]/div[2]/div/h1')
        Dataset_name.append(dataset_name.text)
    except NoSuchElementException:
        Dataset_name.append('-')

    try:
        dataset_type = driver.find_element(By.XPATH,'//*[@class="flex flex-col"]/div[2]/div[2]/div[1]/p')
        Data_type.append(dataset_type.text)
    except NoSuchElementException:
        Data_type.append('-')

    try:
        task = driver.find_element(By.XPATH,'//*[@class="flex flex-col"]/div[2]/div[2]/div[3]/p')
        Task.append(task.text)
    except NoSuchElementException:
        Task.append('-')

    try:
        attribute_type = driver.find_element(By.XPATH,'//*[@class="flex flex-col"]/div[2]/div[2]/div[4]/p')
        Attribute_type.append(attribute_type.text)
    except NoSuchElementException:
        Attribute_type.append('-')

    try:
        no_of_instances = driver.find_element(By.XPATH,'//*[@class="flex flex-col"]/div[2]/div[2]/div[5]/p')
        No_of_instances.append(no_of_instances.text)
    except NoSuchElementException:
        No_of_instances.append('-')

    try:
        no_of_attribute = driver.find_element(By.XPATH,'//*[@class="flex flex-col"]/div[2]/div[2]/div[6]/p')
        No_of_attribute.append(no_of_attribute.text)
    except NoSuchElementException:
        No_of_attribute.append('-')

    try:
        year = driver.find_element(By.XPATH,'//*[@class="flex flex-col"]/div[1]/div[2]/h2')
        Year.append(year.text.replace('Donated on '," "))
    except NoSuchElementException:
        Year.append('-')
```

```python
df = pd.DataFrame({})
df['Data Name'] = Dataset_name 
df['Data Type '] = Data_type
df['Task '] = Task 
df['Attribute Type '] = Attribute_type 
df['No of Instance '] = No_of_instances
df['No of Attributes '] = No_of_attribute
df['Year '] = Year 
df
```
